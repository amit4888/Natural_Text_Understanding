# 2. Tokenization

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([df])
```

```python
print("Total Words:",len(tokenizer.word_index))
```

```python
# here, I am running a loop for texts_to_sequences
input_sequences = []

for sentence in df.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    
    for i in range(1, len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
max_len = max([len(x) for x in input_sequences])
print("Maximum length in a sentence",max_len)
```

# 3. Padding

```python
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len , padding = 'pre')
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
y.shape
```

```python
y = to_categorical(y,num_classes = 3528)
```

```python
y.shape
```

# Build the model

```python
model = Sequential()
model.add(Embedding(input_dim=3528, output_dim=100, input_shape=(57,)))
model.add(SimpleRNN(200))
model.add(Dense(3528,activation = 'softmax'))
```

```python
# Compile the model
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
# Train the model
model.fit(X,y,epochs = 200)
```

# Evaluation

```python
loss, accuracy = model.evaluate(X,y)
print("Test Accuracy:", accuracy)
```
